```python
epoch_count=10000000000000000
batch_size=2
learning_rate=0.0001
learning_decrease=0.99
restore_='result_test.csv'

conf1 = \
  ['trans_conv',{'ksize':[2,2],'stride':[1,1], 'chn':2}]
       
vsm1 = RnnExtModel('vsm1', vsd, conf1)
vsm1.exec_all(epoch_count=epoch_count, batch_size=batch_size, 
                        learning_rate=learning_rate, report=1, show_cnt=0,learning_decrease=learning_decrease,resotre_=restore_)
```
